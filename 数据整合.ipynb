```python
import pandas as pd
import numpy as np
```

# 申万数据处理

```python
sw_1 = pd.read_excel('sw_1.xlsx',index_col=0)
sw_2 = pd.read_excel('sw_2.xlsx',index_col=0).T
sw_3 = pd.read_excel('sw_3.xlsx',index_col=0).T
sw_4 = pd.read_excel('sw_4.xlsx',index_col=0)
```

```python
sw = pd.concat([sw_1,sw_2,sw_3])
sw.index = pd.to_datetime(sw.index)
```

```python
hs300 = pd.read_excel('沪深300指数行情数据.xlsx')
```

```python
sw_1 = sw.merge(hs300,left_index=True,right_on='时间')
```

```python
sw_1.index = sw_1['时间']
```

```python
sw_1.drop('时间',axis=1,inplace=True)
```

```python
sw_1.head()
```

```python
sw_1 = pd.concat([sw_1,sw_4])
```

```python
sw_2 = np.log(1+sw_1.pct_change())
```

```python
sw_2.index = pd.to_datetime(sw_2.index)
```

```python
sw_3 = sw_2.loc['2005-01-31':,:]
```

```python
sw_3.to_csv('sw.csv')
```
